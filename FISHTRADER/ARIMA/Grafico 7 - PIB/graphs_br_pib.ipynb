```python
#importing necessary packages

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
import imageio
import os
```

```python
#função para tratar campo data
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')

#Agora vamos importar nosso arquivo 
df = pd.read_excel('PIB_br.xls', sheet_name='Planilha1')
print(df.head(10))
```

```python
# Definindo as variaveis com as colunas do df
x = df['date']
y1 = df['etanol']
y2 = df['PIB']
#y3 = df_milho['producao']

# Cria uma nova Figura e Axes
fig, ax1 = plt.subplots()
 
# Cria um segundo Axes que compartilha o mesmo eixo x
ax2 = ax1.twinx()
#ax3 = ax2.twinx()

# Plotar dados em cada Axes
ax1.plot(x, y1, 'k-', label='etanol', marker='.', linestyle='dashed')
ax2.plot(x, y2, 'r-', label='PIB', marker='.', linestyle='solid')   #marker='o'
#ax3.plot(x, y3, 'b-', label='milho', marker='.', linestyle='dotted')   #marker='o'

# Definir rótulos dos eixos y
ax1.set_ylabel('Producação de litros de etanol', color='k', )
ax2.set_ylabel('PIB', color='r')
#ax3.set_ylabel('', color='b')

# Incluir legenda
ax1.legend(loc='center left')
ax2.legend(loc='center right')
 
# Mostrar o gráfico
plt.show()
```
