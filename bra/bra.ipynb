```python
# Function to classify salary slabs
def classify_salary(income):
    if income < 5000:
        return 'less than 5k'
    elif 5000 <= income <= 10000:
        return '5k to 10k'
    elif 10000 < income <= 15000:
        return '10k to 15k'
    else:
        return '+15k'
```

```python
# Create the 'salarySlab' column
df['salarySlab'] = df['MonthlyIncome'].apply(classify_salary)
```

```python
# Function to classify age groups
def classify_age(age):
    if 18 <= age <= 25:
        return '18-25'
    elif 26 <= age <= 35:
        return '26-35'
    elif 36 <= age <= 45:
        return '36-45'
    else:
        return '45+'
```

```python

# Create the 'ageGroup' column
df['ageGroup'] = df['Age'].apply(classify_age)
```

```python
df['AttritionRate']
```

```python
# Function to calculate AttritionRate
def calculate_attrition_rate(df):
    attrition_rate = df['AttritionCount'].sum() / df['EmployeeCount'].sum()
    return attrition_rate

# Call the function and store the result in a new column
attrition_rate = calculate_attrition_rate(df)
df['AttritionRate'] = attrition_rate
```

```python
df = df.drop(['AttritionRate'], axis =1)
```

```python
plt.figure(figsize=(7, 5))

def cat_summary(df, col_name, plot=False):
    print(pd.DataFrame({col_name: df[col_name].value_counts(),
                        "Ratio": 100 * df[col_name].value_counts() / len(df)}))

    if plot:
        ax = sns.countplot(x=df[col_name], data=df)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        plt.show(block=True)

for col in cat_cols:
    if df[col].dtypes == "bool":
        df[col] = df[col].astype(int)
        cat_summary(df, col, plot=True)

    else:
        cat_summary(df, col, plot=True)
        print("##########################################")
```

```python
# encoding categorical variables

df = df.copy()

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder() 

for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

# Split dataframe into X and y

```python
X=df.drop(columns='Attrition')
y=df['Attrition']
```

# split dataframe into train and test

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.25 , random_state=42)
```

# Random Forest Classifier 

```python
RF_model = RandomForestClassifier(n_estimators=100)
RF_model.fit(X_train , y_train)
RF_model.score(X_train , y_train)
RF_pred = RF_model.predict(X_test)
print( 'accuracy score: ' ,accuracy_score(y_test , RF_pred))
```

```python
print(classification_report(y_test , RF_pred))
```

```python
cm1 = confusion_matrix(y_test,RF_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm1,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truh')
```
