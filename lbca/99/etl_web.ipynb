```python
import pandas as pd
## SO 
```

```python
# Passo 1: Ler o arquivo Excel
file_path = 'webmotorsG.xlsx'
df = pd.read_excel(file_path)
len(df)
```

```python
# Dicionário com todas as siglas de estados brasileiros e seus nomes por extenso
uf_dict = {
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amapá',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MA': 'Maranhão',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Pará',
    'PB': 'Paraíba',
    'PR': 'Paraná',
    'PE': 'Pernambuco',
    'PI': 'Piauí',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondônia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
}

# Passo 2: Substituir os valores na coluna 'UF' usando o dicionário
df['UF'] = df['UF'].replace(uf_dict)

# Mostrar as primeiras linhas do DataFrame para conferir
print(df['UF'].head())
```

```python
df.head(3)
```

```python
# Passo 2: Identificar as colunas que tenham "data" no nome (case-insensitive)
date_columns = [col for col in df.columns if 'data' in col.lower()]

# Passo 3: Converter essas colunas para o formato datetime
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')  # 'coerce' para valores inválidos

# Mostrar as colunas convertidas para conferência
print(f"Colunas convertidas para formato de data: {date_columns}")
```

```python

# Substituir todas as ocorrências de '***' por espaço vazio
df = df.replace('***', ' ', regex=False)
```

```python
# Salvar de volta como Excel
df.to_excel("webmotors_limpo.xlsx", engine="openpyxl", index=False)
```
