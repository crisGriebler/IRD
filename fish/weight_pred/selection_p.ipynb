```python
import pandas as pd
import joblib  # Usando joblib para carregar o modelo

# Função para carregar o modelo com base no comprimento_Total
def load_model(comprimento_Total: float):
    if comprimento_Total > 22.6:
        model_filename = 'ExtraTreesRegressor_model.pkl'
    else:
        model_filename = 'best_lgbm_model.pkl'
    
    # Carrega o modelo selecionado usando joblib
    model = joblib.load(model_filename)
    
    return model, model_filename
```

```python
def drop_columns(df):
    """
    Drops the specified columns from the DataFrame and returns the resulting DataFrame.
    Also prints the first few rows of the resulting DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    
    Returns:
    pd.DataFrame: The DataFrame after dropping the specified columns.
    """
    columns_to_drop = ['espessura', 'Index','Cluster','cluster']
    df = df.drop(columns=columns_to_drop, inplace=True)
    return df
```

```python
# Carrega o arquivo Excel
data = pd.read_excel('df1.xlsx')
```

```python
drop_columns(data)
data.describe()
```

```python
# Seleciona as primeiras 5 linhas
sample_data = data.head(5)
```

```python
for index, row in sample_data.iterrows():
    comprimento_Total = row['comprimento_Total']
    model, model_filename = load_model(comprimento_Total)
    
    # Preparando os dados de entrada (assumindo que todas as colunas menos a alvo são características)
    features = row.drop(['peso'])  # 'target' é a coluna com o valor real
    features = features.values.reshape(1, -1)  # Ajusta para o formato de entrada do modelo
    
    # Realiza a previsão
    predicted_value = model.predict(features)[0]
    actual_value = row['peso']  # Valor real do target para comparação
    
    # Exibe o modelo selecionado, o valor real e o valor previsto
    print(f"Linha {index + 1}:")
    print(f"  Modelo selecionado: {model_filename}")
    print(f"  Comprimento_Total: {comprimento_Total}")
    print(f"  Valor Real: {actual_value}")
    print(f"  Valor Previsto: {predicted_value}")
    print("-" * 40)
```
