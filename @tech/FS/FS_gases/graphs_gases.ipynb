```python
#função para tratar campo data
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')

#Agora vamos importar nosso arquivo de PIB brasileiro
df = pd.read_csv('emissao.csv')
print(df.head(10))
```

```python
df = df.round(2)
print(df.head(10))
```

```python
# Definindo as variaveis com as colunas do df
x = df['data']
y1 = df['etanol_cana']
y2 = df['etanol_milho']
y3 = df['etanol_total']

# Cria uma nova Figura e Axes
fig, ax1 = plt.subplots()
 
# Cria um segundo Axes que compartilha o mesmo eixo x
#ax2 = ax1.twinx()
#ax3 = ax2.twinx()

# Plotar dados em cada Axes
ax1.plot(x, y3, 'k-', label='etanol-total', marker='o', linestyle='solid')   #marker='o'
ax1.plot(x, y1, 'k-', label='etanol-cana', marker='x', linestyle='dotted')
ax1.plot(x, y2, 'k-', label='etanol-milho', marker='.', linestyle='dashed')   #marker='o'

# Definir rótulos dos eixos y
ax1.set_ylabel('Produção de gases', color='k' )
#ax2.set_ylabel('etanol-milho', color='k')
#ax3.set_ylabel('etanol-total', color='k')

# Incluir legenda
ax1.legend(loc='upper left')
#ax2.legend(loc='center left')
 
# Mostrar o gráfico
plt.show()
```
