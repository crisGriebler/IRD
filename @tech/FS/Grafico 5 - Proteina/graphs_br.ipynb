```python
#importing necessary packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
import imageio
import os
```

```python
#função para tratar campo data
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')

#Agora vamos importar nosso arquivo 
df = pd.read_excel('db_br30_round.xlsx', sheet_name='Sheet1')
print(df.tail(10))
```

```python
df.info()
```

```python
# Definindo as variaveis com as colunas do df
x = df['date']
y1 = df['animal_protein']
y2 = df['ethanol_liter']

# Cria uma nova Figura e Axes
fig, ax1 = plt.subplots()
 
# Cria um segundo Axes que compartilha o mesmo eixo x
ax2 = ax1.twinx()
 
# Plotar dados em cada Axes
ax1.plot(x, y1, 'k-', label='proteina', marker='.', linestyle='dashed')
ax2.plot(x, y2, 'k-', label='etanol', marker='.', linestyle='solid')   #marker='o'
 
# Definir rótulos dos eixos y
ax2.set_ylabel('Produção de etanol mil (milhões de litros)', color='k')
ax1.set_ylabel('Consumo de proteina animal (capita/kg/ano)', color='k')


ax1.legend(loc='upper left')
ax2.legend(loc='lower right')
 
# Mostrar o gráfico
plt.show()
```
