```python
#função para tratar campo data
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')

#Agora vamos importar nosso arquivo 
df1 = pd.read_excel('..\Grafico 5 - Proteina\db_br30.xlsx', sheet_name='Sheet1')
print(df1.head(10))
```

```python
df1.drop(['value'], axis = 1,inplace=True)
```

```python
df1.set_index('date',inplace=True)
```

```python
df1=df1.sort_index(axis=0)
print(df1.tail(10))
```

```python
# Definindo as variaveis com as colunas do df
x = df['Date']
y1 = df1['liter']
y2 = df['PIB']
#y3 = df_milho['producao']

# Cria uma nova Figura e Axes
fig, ax1 = plt.subplots()
 
# Cria um segundo Axes que compartilha o mesmo eixo x
ax2 = ax1.twinx()
#ax3 = ax2.twinx()

# Plotar dados em cada Axes
ax1.plot(x, y1, 'k-', label='etanol', marker='o', linestyle='solid')
ax2.plot(x, y2, 'k-', label='PIB', marker='.', linestyle='dashed')   #marker='o'
#ax3.plot(x, y3, 'b-', label='milho', marker='.', linestyle='dotted')   #marker='o'

# Definir rótulos dos eixos y
ax1.set_ylabel('Producação de litros de etanol', color='k', )
ax2.set_ylabel('PIB', color='k')
#ax3.set_ylabel('', color='b')

# Incluir legenda
ax1.legend(loc='upper left')
ax2.legend(loc='center left')
 
# Mostrar o gráfico
plt.show()
```
