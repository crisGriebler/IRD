```python
#função para tratar campo data
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')

#Agora vamos importar nosso arquivo 
df = pd.read_excel('../../FS/Grafico 6 - Graos/dbmilho_br30.xls', sheet_name='Planilha1')
df_milho=df[37:]
#df_milho.drop(['produtividade', 'area'], axis = 1, inplace=True)
df_milho.set_index('date',inplace=True)
df_milho=df_milho.sort_index(axis=0)
print(df_milho.tail(10))
```

```python
# Definindo as variaveis com as colunas do df
x = df_alcool['date']
y1 = df_alcool['ethanol_liter']
y2 = df_soja['producao']
y3 = df_milho['producao']

# Cria uma nova Figura e Axes
fig, ax1 = plt.subplots()
 
# Cria um segundo Axes que compartilha o mesmo eixo x
ax2 = ax1.twinx()


 
# Plotar dados em cada Axes
ax1.plot(x, y1, 'k-', label='litro de etanol', marker='.', linestyle='solid')
ax2.plot(x, y2, 'k-', label='soja', marker='.', linestyle='dotted')   #marker='o'
ax2.plot(x, y3, 'k-', label='milho', marker='.', linestyle='dashed')   #marker='o'


# Definir rótulos dos eixos y
ax1.set_ylabel('Producação de litros de etanol (milhões de litros)', color='k', )
ax2.set_ylabel('Produção de grãos (toneladas)', color='k')
#ax1.set_ylabel('', color='b')

# Incluir legenda
ax1.legend(loc='lower right')
ax2.legend(loc='upper left')


 
# Mostrar o gráfico
plt.show()
```
